```r
# load the gprofiler library
library(gprofiler2)
```

```r
# load the gene list from txt-file
x <- scan("gene_list.txt", what="", sep="\n")
```

```r
gostres <- gost(query = c(x), organism = "hsapiens",
                ordered_query = FALSE,
                multi_query = FALSE,
                significant = TRUE,
                exclude_iea = FALSE,
                measure_underrepresentation = FALSE,
                evcodes = FALSE,
                user_threshold = 0.05,
                correction_method = "g_SCS",
                domain_scope = "annotated",
                custom_bg = NULL,
                numeric_ns = "",
                sources = NULL,
                as_short_link = FALSE
               )
```

```r
# look at the top 5 results
head(gostres$result)
```

```r
# plot the result overview
p <- gostplot(gostres, capped = FALSE, interactive = FALSE)
p
```
